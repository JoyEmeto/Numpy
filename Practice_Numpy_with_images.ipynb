```python
# Using images to show ways of slicing Numpy array
#load the image using numpy
from skimage import io
photo = io.imread('Image4_002.png')
type(photo)
```

```python
# to check the shape of the photo
photo.shape
```

```python
# to display the image
import matplotlib.pyplot as plt
plt.imshow(photo)
```

```python
#to reverse the photo
plt.imshow(photo[:: -1])
```

```python
#to get a mirror image, get all rows and reverse the columns
plt.imshow(photo[:, ::-1])
```

```python
#to slice a particular part of the image
plt.imshow(photo[250:1500, 2000:3500])
```

```python
# Applying mathematical functions to Numpy
photo
photo_sin = np.sin
print(photo_sin)
```

```python
print(np.sum(photo)) # sum
print(np.prod(photo)) # product
print(np.mean(photo)) # mean
print(np.std(photo)) # standard deviation
print(np.var(photo)) # variance
print(np.min(photo)) # minimum
print(np.max(photo)) # maximum
print(np.argmin(photo)) # arg min
print(np.argmax(photo)) # arg max
# print(np.sqrt(photo)) # square root
```

```python
z = np.array([1, 2, 3, 4, 5])
z
```

```python
# Print true of false for less than 3
z < 3
```

```python
# Print true of false for greater than 3
z > 3
```

```python
# Print list where list is less than 3
z[z < 3]
```

```python
# In the photo, replace where photo is greater than 100 with 225 and where is not greater than 100 with 0

photo_mashed = np.where(photo > 100, 225, 0)
plt.imshow(photo_mashed)
```
